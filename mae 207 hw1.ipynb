```c
void control_motor_loop(Motor_t* motor) {
    while (*(motor->axis_legacy.enable_control)) { //initiates a loop that will continue as long as the enable_control flag is true
        //these 3 if statements check for errors in the motor current measurement, and check if there are errors in the motor operation
        if (osSignalWait(M_SIGNAL_PH_CURRENT_MEAS, PH_CURRENT_MEAS_TIMEOUT).status != osEventSignal) {
            motor->error = ERROR_FOC_MEASUREMENT_TIMEOUT;
            break;
        }

        if (!do_checks(motor))
            break;
        if (!loop_updates(motor))
            break;

        // Only runs if anticogging.calib_anticogging is true; non-blocking
        anti_cogging_calibration(motor);

        // Position control
        // TODO Decide if we want to use encoder or pll position here
        float vel_des = motor->vel_setpoint; //variable for next velocity
        if (motor->control_mode >= CTRL_MODE_POSITION_CONTROL) {
            if (motor->rotor_mode == ROTOR_MODE_SENSORLESS) { //if there is no rotor mode sensor then throw an error
                motor->error = ERROR_POS_CTRL_DURING_SENSORLESS;
                break;
            }
            float pos_err = motor->pos_setpoint - motor->encoder.pll_pos; //difference in position = setpoint - current position
            vel_des += motor->pos_gain * pos_err; //the velocity that you want is variable with units of (1/s) * postiion error
        }

        // Velocity limiting - if the velocity you want is outside the range of the velocity then set it to the max/min velocity 
        float vel_lim = motor->vel_limit;
        if (vel_des > vel_lim) vel_des = vel_lim;
        if (vel_des < -vel_lim) vel_des = -vel_lim;

        // Velocity control
        float Iq = motor->current_setpoint; //current going through motor

        // Anti-cogging is enabled after calibration
        // We get the current position and apply a current feed-forward
        // ensuring that we handle negative encoder positions properly (-1 == motor->encoder.encoder_cpr - 1)
        if (motor->anticogging.use_anticogging) {
            Iq += motor->anticogging.cogging_map[mod(motor->encoder.pll_pos, motor->encoder.encoder_cpr)];
        }

        float v_err = vel_des - get_pll_vel(motor); //velocity error = difference in desired velocity vs. current velocity 
        if (motor->control_mode >= CTRL_MODE_VELOCITY_CONTROL) { //if you're controlling the velocity of the motor
            Iq += motor->vel_gain * v_err; //setting the current equal to [A/(counts/s)]*error in velocity (units counts/s) = end units is Amps
        }

        // Velocity integral action before limiting
        Iq += motor->vel_integrator_current;

        // Apply motor direction correction
        if (motor->rotor_mode == ROTOR_MODE_ENCODER ||
            motor->rotor_mode == ROTOR_MODE_RUN_ENCODER_TEST_SENSORLESS) {
            Iq *= motor->encoder.motor_dir; //making sure the Iq has the same direction as the motor is currently rotating in
        }

        // Current limiting, don't allow current to get above max allowed current
        float Ilim = MACRO_MIN(motor->current_control.current_lim, motor->current_control.max_allowed_current);
        bool limited = false; //variable becomes true if we are trying to set Iq greater than the current limit
        if (Iq > Ilim) { //if the set current is greater than the max allowed current
            limited = true;
            Iq = Ilim; //set Iq equal to the max allowed current
        }
        if (Iq < -Ilim) { //same but for set current less than minimum allowed negative current
            limited = true;
            Iq = -Ilim;
        }

        // Velocity integrator (behaviour dependent on limiting) - 
        if (motor->control_mode < CTRL_MODE_VELOCITY_CONTROL) {
            // reset integral if not in use
            motor->vel_integrator_current = 0.0f;
        } else {
            if (limited) { //if you have a limiting current
                // TODO make decayfactor configurable
                motor->vel_integrator_current *= 0.99f; //set your integrator to be decaying
            } else { //otherwise set your integrator current to gain*period*error in velocity
                motor->vel_integrator_current += (motor->vel_integrator_gain * current_meas_period) * v_err;
            }
        }

        // Execute current command
        if (motor->motor_type == MOTOR_TYPE_HIGH_CURRENT) { //if you have a high current motor
            if(!FOC_current(motor, 0.0f, Iq)){ //set the current of your motor to Iq
                //if the FOC_current function returns 0 then there's an error and you break
                break; // in case of error exit loop, motor->error has been set by FOC_current
            }
        } else if (motor->motor_type == MOTOR_TYPE_GIMBAL) { //if you have a gimbal motor
            //In gimbal motor mode, current is reinterptreted as voltage.
            if(!FOC_voltage(motor, 0.0f, Iq)){
                break; // in case of error exit loop, motor->error has been set by FOC_voltage
            }
        } else { //if your motor type is not high current or gimbal then it's not supported and you get an error
            motor->error = ERROR_NOT_IMPLEMENTED_MOTOR_TYPE;
            break;
        }

        ++(motor->loop_counter); //increment the loop counter
    }

    //We are exiting control, reset Ibus, and update brake current
    motor->current_control.Ibus = 0.0f;
    update_brake_current();
}
```
